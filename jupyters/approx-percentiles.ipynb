```python
import matplotlib.pyplot as plt
import numpy as np

# Generate bimodal data between 1 and 50
data = np.concatenate((np.random.normal(loc=15, scale=5, size=500),
                        np.random.normal(loc=35, scale=5, size=500)))

# Define the color scheme
colors = ['#e5f5e0', '#a1d99b', '#31a354', '#006d2c']

# Create the histogram plot
fig, ax = plt.subplots(figsize=(8, 5))

# Set the number of bins and plot the histogram with spacing between bars
num_bins = 10
n, bins, patches = ax.hist(data, bins=num_bins, edgecolor='white', linewidth=1.2, alpha=0.8, color=colors[1])

# Set the y-axis to show total count of values
ax.set_ylabel('Count', fontsize=14)

# Add a title and axis labels with appealing font
title_font = {'fontname':'Helvetica', 'size':18, 'fontweight':'bold'}
axis_font = {'fontname':'Helvetica', 'size':14}
ax.set_title('Histogram of Bimodal Distribution', **title_font)
ax.set_xlabel('Value', **axis_font)

# Remove the grid and set font properties
ax.grid(False)
ax.tick_params(axis='x', labelsize=12, length=0)
ax.tick_params(axis='y', labelsize=12)

# Add annotations for the starting count of each bin
for i in range(len(n)):
    ax.annotate(str(int(n[i])), xy=(bins[i], 0), xytext=(bins[i]+1, 20),
                fontsize=10, fontweight='bold', ha='center', va='bottom',
                arrowprops=dict(arrowstyle='-[, widthB=1.5, lengthB=0.8', lw=1.2),
                color=colors[2])

# Set the color scheme for the histogram bars and annotations
for i in range(len(patches)):
    patches[i].set_facecolor(colors[1])
    patches[i].set_edgecolor(colors[3])
    patches[i].set_linewidth(1.2)

# Show the plot
plt.show()
```

```python
num_bins = 50
from scipy import stats

# Define the color scheme
colors = ['#e5f5e0', '#a1d99b', '#31a354', '#006d2c']

# Calculate the 30th percentile value and its bin
percentile_val = np.percentile(data, 25)
bin_edges = np.linspace(0, 50, num=num_bins)
percentile_bin = np.searchsorted(bin_edges, percentile_val, side='right') - 1

# Create the histogram plot
fig, ax = plt.subplots(figsize=(7, 4))

# Set the number of bins and plot the histogram with spacing between bars
n, bins, patches = ax.hist(data, align='right', bins=num_bins, edgecolor='white', linewidth=1.2, alpha=0.8, color=colors[1])
# Create a PDF line
pdf = stats.norm.pdf(bins, np.mean(data), np.std(data))
plt.plot(bins, pdf, color='red', linestyle='--')

# Create a CDF line
cdf = stats.norm.cdf(bins, np.mean(data), np.std(data))
plt.plot(bins, cdf, color='blue', linestyle='-')

# Highlight the bin containing the 30th percentile value
patches[percentile_bin].set_facecolor(colors[2])
patches[percentile_bin].set_edgecolor(colors[3])
patches[percentile_bin].set_linewidth(1.5)

# Set the y-axis to show total count of values
ax.set_ylabel('Count', fontsize=14)

# Add a title and axis labels with appealing font
title_font = {'fontname':'Helvetica', 'size':16, 'fontweight':'bold'}
axis_font = {'fontname':'Helvetica', 'size':14}
ax.set_title('Buckets', **title_font)
ax.set_xlabel('Value', **axis_font)

# Remove the grid and set font properties
ax.grid(False)
ax.tick_params(axis='x', labelsize=12, length=0)
ax.tick_params(axis='y', labelsize=12)

# Set the color scheme for the histogram bars and annotations
for i in range(len(patches)):
    patches[i].set_facecolor(colors[1])
    patches[i].set_edgecolor(colors[3])
    patches[i].set_linewidth(1.2)

# Show the plot
plt.show()
fig.savefig('full-histogram.png', dpi=600, bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats

# Create a histogram
n, bins, patches = plt.hist(data, bins=30, density=True, alpha=0.5)

# Create a PDF line
pdf = stats.norm.pdf(bins, np.mean(data), np.std(data))
plt.plot(bins, pdf, color='red', linestyle='--')

# Create a CDF line
cdf = stats.norm.cdf(bins, np.mean(data), np.std(data))
ax2 = plt.twinx()
ax2.plot(bins, cdf, color='blue', linestyle='-', alpha=0.8)

# Set the title and labels
plt.title('CDF vs. PDF Plot')
plt.xlabel('Data')
plt.ylabel('Probability')

# Set the font
plt.rcParams['font.family'] = 'Arial'

# Set the y-axis limits for the PDF
plt.ylim(0, np.max(pdf)*1.2)

# Set the y-axis limits for the CDF
ax2.set_ylim(0, 1)

# Add a legend
plt.legend(['PDF', 'CDF'])

# Show the plot
plt.show()
```
